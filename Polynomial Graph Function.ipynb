```python
## Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np

## Defining the function
def polynomial_graph(degree=2, intercept=0, X_lim=(-10, 10)):
    
    ## Asking user if they want to insert the constant term for each degree
    user_input = input("Do you like to insert the constant term for each degree? (by default 1) (y/n): ")
    
    ## Defining constants to store values, defaulting each to 1
    constant = [int(input(f"Enter constant for degree {degree - i}: ") if user_input == "y" else 1) for i in range(degree)]
    
    ## Defining X range
    X = np.linspace(X_lim[0], X_lim[1], 100)
    
    ## Initializing Y and Y_equation
    Y = 0
    Y_equation = ""
    
    ## Creating equation of the polynomial line
    for i in range(len(constant)):
        if constant[i] == 1:
            Y_equation += f'X^{degree - i} + '
            Y += constant[i] * (X ** (degree - i))
        elif constant[i] == 0:
            Y_equation += f''
        else:
            Y += constant[i] * (X ** (degree - i))
            Y_equation += f'{constant[i]}X^{degree - i} + '
    
    ## Adding intercept part in the equation
    if intercept != 0:
        Y += intercept
        Y_equation += f'{intercept}'
    else:
        Y_equation = Y_equation.rstrip('+ ')

    ## Finding min and max points
    min_index = np.argmin(Y)
    max_index = np.argmax(Y)
    min_point = (X[min_index], Y[min_index])
    max_point = (X[max_index], Y[max_index])

    ## Plotting the graph
    plt.figure(figsize=(10, 8))
    plt.plot(X, Y, color='red', label=f'Y = {Y_equation}')
    plt.grid(True, linestyle='--')
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')

    ## Annotating min and max points
    plt.scatter(*min_point, color='blue', s=50)
    plt.scatter(*max_point, color='green', s=50)
    plt.text(min_point[0]+2.8, min_point[1], f'({min_point[0]:.2f}, {min_point[1]:.2f})', color='blue', ha='center', fontsize=10, fontweight='bold')
    plt.text(max_point[0]-2.8, max_point[1], f'({max_point[0]:.2f}, {max_point[1]:.2f})', color='green', ha='center', fontsize=10, fontweight='bold')
    
    ## Adding equation as the title
    plt.title(f'Equation of Line : $Y = {Y_equation}$', fontsize=15, fontweight='bold')
    plt.show()

## Calling the function
polynomial_graph(3)
```
